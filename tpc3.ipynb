```
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#inicialização datasets
data = datasets.california_housing.fetch_california_housing()
#criacao dataFrame
dframe = pd.DataFrame(data = data.data, columns=data.feature_names)

'done'
```

```
#1) Aggregation
dframe.agg(["min","max","mean"])
```

```
#2) Sampling
dframe.sample(4)
```

```
#3) Dimensionality Reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
pca.fit(dframe)
columns = ['pca.%i' % i for i in range(4)]
df_pca = pd.DataFrame(pca.transform(dframe), columns=columns, index=dframe.index)

df_pca.head()
```

```
#4) Feature Subset Selection
dframe[['HouseAge','Population','AveOccup']][0:5]
```

```
#5) Feature Creation
dframe['AveRooms/Population']= dframe['AveRooms']/ dframe['Population']

dframe.head()
```

```
#6) Discretization and Binarization
binary = dframe.Population[0:5]
pd.get_dummies(binary)
```

```
#8) Attribute Transformation
dframe['NewPopulation'] = dframe['Population'].transform(lambda x : x+20)
dframe.head()
```
