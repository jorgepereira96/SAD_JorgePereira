```
def solution(S):
  #se a string estiver vazia retorna 1
  if not S:
    return 1
  #enquanto a string tiver pelo menos 1 caractere, os caracteres iram ser substituidos por vazio
  while len(S)!=0:
    ex=S
    S=S.replace('()','').replace('[]','').replace('{}','')
    if ex is S:
      return 0
    
  return 1

print(solution('{[()()]}'))
print(solution('([) ()]'))
```
